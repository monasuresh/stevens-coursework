# 1. Explain the following concepts

# Supervised Learning

Supervised learning is a fundamental category in machine learning where the training data supplied to the system comprises known solutions, often referred to as labels. In this process, a model is presented with inputs along with their corresponding correct outputs. There are primarily two types of tasks within supervised learning: classification and regression.

In classification tasks, the machine learning system is exposed to numerous instances, each accompanied by its associated class. The primary objective here is for the system to acquire the ability to classify new, unseen data accurately.

On the other hand, in regression tasks, the goal is to predict a numerical target value based on a given set of features, which are referred to as predictors. To train the system effectively, it is essential to provide it with a substantial number of examples that encompass both the predictors and their corresponding labels.

Various supervised learning algorithms are employed in these tasks, each with its unique characteristics. Some of the commonly used supervised learning algorithms include k-Nearest Neighbors, Linear Regression, Logistic Regression, Support Vector Machines, Decision Trees, Random Forests, and Neural Networks.


# Unsupervised Learning

Unsupervised learning is a category of machine learning in which the training dataset lacks labeled responses, and the system endeavors to acquire knowledge independently without the guidance of a teacher. Within unsupervised learning, there exist various algorithms designed for specific tasks, including clustering, anomaly detection, novelty detection, visualization, dimensionality reduction, and association rule learning.

For instance, clustering algorithms are employed to group unlabelled data based on their similarities or dissimilarities, effectively organizing data points into distinct clusters. Visualization techniques are utilized to transform unlabelled data into 2D or 3D representations, making it easier to visualize and analyze. This transformation is achieved through dimensionality reduction, a process that reduces the number of features in the dataset while retaining its essential information.

Anomaly detection and novelty detection are also integral components of unsupervised learning, as they involve identifying unusual patterns or data points within unlabelled data. In both cases, information regarding which data points are anomalies or outliers is not provided, necessitating the detection of anomalies directly from the data itself. Association rule learning, which seeks to uncover relationships between items within large datasets, is another example that operates without the need for labeled data. If associations are detected among data points, they are grouped together accordingly.


# Online Learning

Online learning is a distinct method for categorizing machine learning systems. In this approach, machine learning systems undergo gradual and continuous training over a period of time, as opposed to being trained all at once. During this process, data instances are presented to the system individually or in small groups known as mini-batches. An essential element of online learning is the learning rate, which plays a pivotal role in the ml system's behavior.

The learning rate determines the speed at which the system grasps new data. A high learning rate enables rapid adaptation to new information but may result in the system forgetting old data quickly. Conversely, a low learning rate hinders swift data acquisition but makes the system less responsive to noise and slower to adapt to shifts in data patterns.

Online learning excels when dealing with continuously incoming data and necessitates rapid system adaptation to changing circumstances. Importantly, once the system has acquired knowledge about a specific data instance, it no longer requires that data, allowing for efficient data storage management through discarding unnecessary information.

Moreover, online learning is particularly advantageous when handling extensive datasets that exceed the memory capacity of a single machine, as it processes data incrementally, eliminating the need to load the entire dataset into memory at once.

Nonetheless, online learning presents challenges, notably in terms of performance degradation when the system is exposed to erroneous or low-quality data. To address this issue, continuous monitoring of the system is essential to identify and rectify performance issues promptly.


# Batch Learning

Batch learning is a specific approach to training machine learning systems, wherein the entire dataset is used for training simultaneously, as opposed to using smaller batches or incremental updates. Due to its computational demands, batch learning is typically carried out offline. In this process, ml systems are trained comprehensively and then deployed into a production environment, where they cease to learn further. Instead, they apply the knowledge they have acquired during training.

To enable a batch learning system to accommodate new data, it necessitates a complete retraining process that encompasses both the new data and the existing dataset. This entails replacing the old system with the new one. Automation plays a crucial role in facilitating this adaptation of batch learning systems to new data.

However, retraining a system with new data is a time-consuming process, and as a result, new systems are often trained on a scheduled basis, such as every 24 hours or on a weekly cadence. Given that batch learning requires significant computational resources, including CPU, memory, and I/O, it may not be feasible for handling large datasets. Achieving autonomous learning within a batch learning framework can also be considerably challenging.


# Model-based Learning

Model-based learning involves making predictions by constructing a model based on a given set of examples. To do this, you first choose a model from a selection of potential models and specify the values for its parameters. These parameter values are selected to optimize the model's performance.

The performance of a model can be assessed using two types of functions: a utility function, which quantifies how well the model performs, and a cost function, which quantifies how poorly it performs. To train the model, you provide it with training examples and adjust its parameters to make it fit the data points as closely as possible. Once you have achieved the best fit, you can use the model to generate predictions.

Effective models yield accurate predictions, but if the predictions are inaccurate, several steps can be taken to improve the model. These include adding additional attributes, using higher-quality training data, or selecting a more robust model with enhanced capabilities.

# Instance-based Learning

In instance-based learning, the system memorizes specific examples or instances from the training data to make predictions for new, unseen data. The system doesn't build an explicit model. Rather the training data is stored as a set of instances. The system generalizes to new examples by using a similarity measure to compare new examples/instances to instances with labels stored in the training data. Because instance-based learning systems simply memorize the training instances and don't learn a model, they don't generalize to new data as model-based learning systems and their predications aren't as reliable. Also because there isn't a model to be trained, predications can be slow because the system would need to compare similarity with existing instances.

```python
#Import Statements
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid') 
import numpy as np
```

# Training Data

```python
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

# 2. Program to estimate the parameters for an unknown polynomial

```python
import numpy as np
import matplotlib.pyplot as plt

def fit_polynomial(x, y, degree):
    # Turn the input arrays into vectors
    vector_x = x.ravel()
    vector_y = y.ravel()

    # Fit a polynomial of the specified degree
    coefficients = np.polyfit(vector_x, vector_y, degree)
    p = np.poly1d(coefficients)

    # Sort x values
    sorted_x = np.sort(vector_x)

    # Get the y values for the sorted polynomial
    y_curve = p(sorted_x)

    polyData = {
        "p": p,
        "vector_x": vector_x,
        "vector_y": vector_y,
        "sorted_x": sorted_x,
        "y_curve": y_curve
    }

    return polyData
```

# Q1. Plot the noisy data and polynomial

```python
polyData = fit_polynomial(x, y, 4)

plt.title('Polynomial of m = 4 and Noisy Data')
plt.plot(polyData["vector_x"], polyData["vector_y"], 'co', label='Noisy Data')
plt.plot(polyData["sorted_x"], polyData["y_curve"], 'm-', label='Polynomial Curve')
plt.legend()
plt.show()
```

# Q2. Plot MSE versus m, for m = 1, 2, 3, 4, 5, 6, 7, 8

```python
mse_array = []
degrees = [1, 2, 3, 4, 5, 6, 7, 8]
for degree in degrees:
    polyData = fit_polynomial(x, y, degree)
    poly = polyData["p"]
    vector_x = polyData["vector_x"]
    vector_y = polyData["vector_y"]
    
    mse = sum((vector_y - poly(vector_x))**2) / number_of_samples
    mse_array.append(mse)
    
plt.title('MSE vs. M')
plt.xlabel('M (Degree)')
plt.ylabel('Mean Squared Error')
plt.plot(degrees, mse_array)
plt.show()

for i in range(len(mse_array)):
    print('m = ' + str(i + 1) + ', ' + str(mse_array[i]))
```

The best value for m in this context is 3. To decide on the best m, we consider two things: Mean Squared Error (MSE) and the risk of overfitting.

MSE measures how closely our recovered polynomial's predictions match the actual data points. We want MSE to be low because this indicates a good fit between our recovered parameters and the data. However, the lowest MSE value is 8, but that doesn't mean we should choose m as 8.

Choosing m as 8 might result in an excellent fit to the training data but it could lead to overfitting. Overfitting means the model becomes too complex and fits the training data noise rather than the underlying pattern. This can make the model perform poorly on new, unseen data.

So the better choice for m is 3 because it creates a simpler model. Simpler models are less likely to overfit and are more likely to generalize to and perform well on new data. Looking at a graph and at the MSE values printed below the graph, you can see that starting from m = 3 and going higher, the MSE values don't change significantly. This suggests that increasing m beyond 3 doesn't provide much benefit, and it's better to stick with a simpler model to reduce the risk of overfitting.

# Q3. Change variable noise_scale to 150, 200, 400, 600, 1000

```python
for noise_scale in [150, 200, 400, 600, 1000]:
    number_of_samples = 50
    x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
    y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    
    polyData = fit_polynomial(x, y, 3)
    
    plt.title(f'Noise scale {noise_scale} plotted against polynomial with m = 3')
    plt.plot(polyData["vector_x"], polyData["vector_y"], 'co', label='Noisy Data')
    plt.plot(polyData["sorted_x"], polyData["y_curve"], 'm-', label='Polynomial Curve')
    plt.legend()
    plt.show()
```

After repeatedly running the Jupyter notebook, I observed some trends related to the accuracy of polynomial paramters as they relate to the noise scale. As the level of noise in the data increases, the precision of our parameter estimates diminishes. This phenomenon occurs due to the presence of greater noise, which introduces more random and meaningless information as well as inaccuracies into the dataset. Consequently, it becomes more challenging to accurately discern the underlying data pattern.

Consequently, when there is a higher degree of noise, it becomes increasingly difficult to fit a curve that closely aligns with the data points. In the provided graphs, it is evident that with a low noise level (150), the curve more closely tracks the data points because of less scattering of the points. With a noise scale of 200, the model is able to capture the overall pattern in the data, although not as well as when the noise scale is 150. The data points start to become more scattered. As the noise scale rises, the polynomial fit to the training data becomes less precise. For instance, when compared to a noise scale of 150 or 200, a noise scale of 400 leads to a significantly less accurate fit and a flatter curve. This is observable by the increased scattering and noise in the data points, which prevents the polynomial from capturing the underlying pattern. As the noise scale escalates to 600, the data points become even more dispersed, resulting in underfitting due to a diminished trend or pattern in the data. The model struggles to identify or learn a consistent pattern within the dataset, resulting in a noticeably flatter plotted polynomial. When the noise level is high (1000), the recovered polynomial deviates substantially from the training data, and the data points exhibit even greater dispersion. 

In summary, when more noise was introduced into the training data, it had a detrimental effect on the accuracy of the polynomial parameters that were generated. Conversely, when the training data had less noise, the accuracy of the resulting parameters improved. This improvement can be attributed to the fact that with less noise, it becomes easier to discern meaningful patterns within the data.

# Q4. Change variable number_of_samples to 40, 30, 20, 10

```python
for number_of_samples in [40, 30, 20, 10]:
    noise_scale = 100
    x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
    y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    
    polyData = fit_polynomial(x, y, 3)
    
    plt.title(f'{number_of_samples} samples plotted against polynomial with m = 3')
    plt.plot(polyData["vector_x"], polyData["vector_y"], 'co', label='Noisy Data')
    plt.plot(polyData["sorted_x"], polyData["y_curve"], 'm-', label='Polynomial Curve')
    plt.legend()
    plt.show()
```

Similarly, with a sample size of 20, the polynomial curve exhibited jaggedness instead of a smooth curve. This suggests that the model struggles to capture the overall data trend and instead also attempts to fit the noise within the data. With a limited number of data points, the noise can have a significant influence on the estimated coefficients, making them less accurate. Also depending on how the model performs on new data when compared to how it performs on the training data, overfitting may occur.

Conversely, when the sample size increases to 30, the polynomial curve generally becomes smoother in comparison to the 10 and 20 sample sizes. With a sample size of 40, the curve consistently appeared at its smoothest and predicted the overall trend and pattern present in the training data. This occurs because the polynomial model starts to generalize from the data, rather than fitting the noise. When the model can effectively generalize, it tends to yield more accurate predictions and successfully captures the underlying data pattern.

In summary, having a larger number of samples may result in better predictions for new data, whereas a limited sample size often causes models to attempt to fit to noise rather than generalize to the data, leading to poorer performance when tested with new data.
